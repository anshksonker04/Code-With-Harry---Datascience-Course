```python
import json

def load_data(filename):
    with open(filename, 'r') as f:
        data= json.load(f)
    return data
```

```python
def display_data(data):
    user_lookup= {user['id']:user['name'] for user in data['users']}
    page_lookup= {page['id']:page['name'] for page in data['pages']}
    print('Users and connections are:-')

    for user in data['users']:
        friends_names= [user_lookup[friend_id] for friend_id in user['friends']]
        page_names= [page_lookup[page_id] for page_id in user['liked_pages']]

        print(f"ID:{user['id']} - {user['name']} is friends with {friends_names} and liked pages are {page_names}")
    print('\nPage information is displayed below:')
    for page in data['pages']:
        print(f"ID:{page['id']} - {page['name']}")
```

```python
data= load_data('data.json')
display_data(data)
```
