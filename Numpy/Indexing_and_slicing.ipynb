```python
# fancy indexing -> select multiple elements
arr= np.array([12,3,45,65,4,66,32,47,90,87])
arr[[1,2,3]]
```

```python
# Boolean Masking for filtering data
arr= np.array([12,3,45,65,4,66,32,47,90,87])
mask = arr > 34
arr[mask]
```
