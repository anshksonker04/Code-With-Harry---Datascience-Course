```python
# Using Python lists
import time

size= 1_000_000_0
l1= list(range(size))
l2= list(range(size))

start= time.time()
add= [x+y for x,y in zip(l1,l2)]
end= time.time()
print(add[0:10])
print(end- start)
```

```python
# Using Numpy Arrays
import numpy as np
import time

size= 1_000_000_0
l1= np.array(range(size))
l2= np.array(range(size))

start= time.time()
add= l1+l2
end= time.time()
print(add)
print(end- start)
```
