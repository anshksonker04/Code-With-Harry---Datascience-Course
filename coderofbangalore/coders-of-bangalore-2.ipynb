```python
with open('initialdata.txt', encoding= 'utf-8') as f:
    data= f.read()
```

```python
chunks= data.split('\n\n')
chunks= [c for c in chunks if len(c)>3]
```

```python
def parse_chunk(chunk):
    try:
        chunk= chunk.strip()
        sep_chunk= chunk.split('\n')
        username= sep_chunk[0]
        no_of_posts= int(sep_chunk[1].split(" post")[0].replace(",", ""))
        no_of_followers= float(sep_chunk[2].split(" follower")[0].replace(",", "").replace("K", "").replace("M", ""))
        if("K" in sep_chunk[2]):
            no_of_followers= int(no_of_followers * 1000)
        elif("M" in sep_chunk[2]):
            no_of_followers= int(no_of_followers * 1000000)
        else:
            no_of_followers= int(no_of_followers)
            
        no_of_following= int(sep_chunk[3].split(" following")[0].replace(",", "").replace("K", "").replace("M", ""))
        if("K" in sep_chunk[3]):
            no_of_following= int(no_of_following * 1000)
        elif("M" in sep_chunk[3]):
            no_of_following= int(no_of_following * 1000000)
        else:
            no_of_following= int(no_of_following)
            
        real_name= sep_chunk[4]
        if(len(sep_chunk)>5):
            type_of_page= sep_chunk[5]
            bio= "\n".join(sep_chunk[6:])
        else:
            type_of_page= "Unknown"
            bio= ""
        # print(username, no_of_posts, no_of_followers, no_of_following, real_name, type_of_page, bio, sep='\n')
        return {"username": username, "posts": no_of_posts, "followers": no_of_followers, "following": no_of_following,
                "name": real_name, "page": type_of_page, "bio": bio}
    except Exception as e:
        print(e, chunk)
```

```python
parse_chunk(chunks[1])
```

```python
all_chunks= []
for chunk in chunks:
    parsed_chunk= parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
print(all_chunks)    
```

```python
import json
s= json.dumps(all_chunks, indent=4)
print(s)
```

```python
# Username with the maximum posts --> self
all_posts= []
for user in all_chunks:
    all_posts.append(user["posts"])

max_posts= max(all_posts)
max_posts

for user in all_chunks:
    if(user['posts']== max_posts):
        max_posts_holder= [user['username'], user['name']]
        break;
print(f"Page with the maximum number of posts is {max_posts_holder[1]}, username= {max_posts_holder[0]} with {max_posts} posts")
```

```python
# Person with the maximum followers --> self
followers_lookup= [user['followers'] for user in all_chunks]
max_followers= max(followers_lookup)

max_followers_holder= [user for user in all_chunks if(user['followers']==max_followers)]
print(f"Person with the most followers is {max_followers_holder[0]['name']} and its username is {max_followers_holder[0]['username']} with {max_followers} followers")
```

```python
# Person who follows the maximum people --> self
following_lookup= [user['following'] for user in all_chunks]
max_following= max(following_lookup)

max_following_holder= [user for user in all_chunks if(user['following']==max_following)]
print(f"Person with the most followers is {max_following_holder[0]['name']} and its username is {max_following_holder[0]['username']} with {max_following} following")
```
